```python
#camper script
class Camper:

    def __init__(self, name):

        self.name = name
        self.hometown = ' '
        self.age = int()
        self.experience_level = ['beginner', 'intermediate', 'expert']
        self.destination = ''
```

```python
#camping equipment script
class Tent:

    def __init__(self, name):

        self.name = name
        self.brand = ''
        self.dimensions = ''
        self.seasons = 3,4
        self.capacity = ''
        self.style = cabin, dome

class backpack:

    def __init__(self, name):

        self.name = name
        self.brand = ''
        self.pockets = ''

class campingLocation:

    def __init__(self, location):

        self.location = location
    

    def set_location(self):
        
        #get the coordinates for camping locations
        df = pd.DataFrame({'name': ['durango, colorado', 'bigfork, montana', 'eureka springs, arkansas']})

        from geopy.geocoders import Nominatim
        geolocator = Nominatim(user_agent="tellergw@gmail.com")

        from geopy.extra.rate_limiter import RateLimiter
        geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
        
        df['location'] = df['name'].apply(geocode)
        df['point'] = df['location'].apply(lambda loc: tuple(loc.point) if loc else None)
        df[['latitude','longitude','O']] = pd.DataFrame(df['point'].tolist(), index=df.index)
        df = df.drop('O', axis = 1)
        
        #get away from hard coding this part
        durango_point = df.loc[0,'latitude':'longitude']
        bigfork_point = df.loc[1,'latitude':'longitude']
        eureka_point = df.loc[2,'latitude':'longitude']
        
        durango_point = list(durango_point)
        bigfork_point = list(bigfork_point)
        eureka_point = list(eureka_point)

        if self.location == 'durango, colorado':
            self.location = ",".join([str(i) for i in durango_point])
        elif self.location == 'bigfork, montana':
            self.location= ",".join([str(i) for i in bigfork_point])
        elif self.location == 'eureka springs, arkansas':
            self.location = ",".join([str(i) for i in eureka_point])
        else:
            print('Sorry, campsites are not available.')
     
        
    def set_forecast(self):
        
        url = "https://api.tomorrow.io/v4/timelines"

        querystring = {
        # "location":"37.2395367, -107.82158179169292",
        "location":self.location,
        "fields":["temperature", "precipitationProbability"],
        "units":"imperial",
        "timesteps":"1d",
        "apikey":"7Eh44e43xZb8o2k5KJEs9WBdUgNBcQl5"}

        response = requests.request("GET", url, params=querystring)
#         print(response.text)
        print("Weather Forecast")
        print("================")

        location_results = response.json()['data']['timelines'][0]['intervals']
        
        for daily_result in location_results:
            date = daily_result['startTime']
            temp = round(daily_result['values']['temperature'])
            precipitationProbability = (daily_result['values']['precipitationProbability'])
            
            print("On",date,"it will be", temp, "F", 'the chance of precipitation is :', precipitationProbability,'%')
```

```python
#create a camping checklist
from tkinter import *

class checklist:
    
    def __init__(self,checklist_item_1, checklist_item_2, checklist_item_3):
        self.checklist_item_1 = checklist_item_1
        self.checklist_item_2 = checklist_item_2
        self.checklist_item_3 = checklist_item_3
        
    def make_gui(self):
        
        root = Tk()
        root.geometry("400x200")

        w = Label(root, text ='Camping Checklist:', font = "50")
        w.pack()

        Checkbutton1 = IntVar()
        Checkbutton2 = IntVar()
        Checkbutton3 = IntVar()

        Button1 = Checkbutton(root, text = self.checklist_item_1,
                                variable = Checkbutton1,
                                onvalue = 1,
                                offvalue = 0,
                                height = 2,
                                width = 20)

        Button2 = Checkbutton(root, text = self.checklist_item_2,
                                variable = Checkbutton2,
                                onvalue = 1,
                                offvalue = 0,
                                height = 2,
                                width = 20)

        Button3 = Checkbutton(root, text = self.checklist_item_3,
                                variable = Checkbutton3,
                                onvalue = 1,
                                offvalue = 0,
                                height = 2,
                                width = 20)

        Button1.pack()
        Button2.pack()
        Button3.pack()

        mainloop()
```

```python
# https://pypi.org/project/camply/
#searches through recreation.gov for available campsites
    
from datetime import datetime
import logging
from typing import List

from camply.containers import AvailableCampsite, SearchWindow
from camply.search import SearchRecreationDotGov

class campSite_checker:
    
    def __init__(self,campsite_location):

        self.campsite_location = campsite_location

    def set_campsite(self):
        
        if self.campsite_location == 'durango, colorado':
            self.campsite_location = 1054 #san juan national forest
        elif self.campsite_location == 'bigfork, montana':
            self.campsite_location = 2725 # Glacier Ntl Park
        elif self.campsite_location == 'eureka springs, arkansas':
            self.campsite_location = 530 #Beaver Lake
        else:
            print('Sorry, campsites are not available.')


        logging.basicConfig(format="%(asctime)s [%(levelname)8s]: %(message)s",
                        level=logging.INFO)

        for_next_month = SearchWindow(start_date=datetime(year=2022, month=9, day=1),
                                 end_date=datetime(year=2022, month=9, day=15))
        camping_finder = SearchRecreationDotGov(search_window=for_next_month,
                                            recreation_area = self.campsite_location,  
                                            weekends_only=False,
                                            nights=1)
        matches: List[AvailableCampsite] = camping_finder.get_matching_campsites(log=True, verbose=True, 
                                                                                 continuous=False)
```
